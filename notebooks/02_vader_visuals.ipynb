```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/qasim/Documents/Analyzer Internship Project/Summer_IST495/data/vader_comparison_results.csv')
df.head()
```

```python
# Count the sentiment labels
sentiment_counts = df['vader_sentiment'].value_counts()
print(sentiment_counts)
```

```python
# Plot bar chart
sentiment_counts.plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('VADER Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Headlines')
plt.tight_layout()
plt.show()
```

VADER Sentiment Distribution Bar Chart Notes:

Shows how VADER classified the overall tone of each headline as positive, neutral, or negative.
Some headlines labeled as positive or negative by VADER did not match manual expectations, highlighting the limitations of rule-based scoring.
Distribution helps identify if VADER is biased toward any particular sentiment.

```python
# Count matches
match_counts = df['match'].value_counts()

# Plot pie chart
match_counts.plot(kind='pie', labels=['Match', 'Mismatch'], autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('VADER vs. Manual Sentiment Match Rate')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

Pie Chart - VADER vs. Manual Sentiment Match:

Visualizes the proportion of headlines where VADER's predicted sentiment matched vs. mismatched manual labels.
Achieved an overall accuracy of 60%, meaning 40% of predictions did not align with human judgment.
Highlights the need to review edge cases and refine the scoring logic.
This chart reinforces the idea that while VADER is helpful, it shouldnt be fully relied on without lots of customization.
