```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

# applying sentiment scoring to the first 100 reviews
df = pd.read_csv('/Users/qasim/Documents/Analyzer Internship Project/Summer_IST495/data/Reviews.csv')[['Score', 'Text']].dropna().head(100)
df['vader_score'] = df['Text'].apply(lambda text: sia.polarity_scores(str(text))['compound'])


# Apply VADER and print results
for i, row in df.iterrows():
    text = str(row['Text'])
    score = row['Score']
    vader_result = sia.polarity_scores(text)
    compound = vader_result['compound']
    
    print(f" Review Score: {score}")
    print(f" VADER Compound: {compound}")
    print(f" Text: {text[:300]}...")  # Shows only 300 characters
    print("-" * 80)
    print() #formatting purposes
```

I used nltk.sentiment.vader.SentimentIntensityAnalyzer to generate a compound sentiment score for each review. Now, I will create an accuracy comparison.

```python
def label_vader(compound):
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

def label_star(score):
    if score >= 4:
        return 'positive'
    elif score == 3:
        return 'neutral'
    else:
        return 'negative'

df['vader_label'] = df['vader_score'].apply(label_vader)
df['true_label'] = df['Score'].apply(label_star)

# Accuracy
accuracy = (df['vader_label'] == df['true_label']).mean()
print(f"VADER Accuracy vs Review Score: {accuracy:.2f}")
```
